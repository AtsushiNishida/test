```python
import GPy
import GPyOpt
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Object to optimize
# function is unknown for your application, but you just need to set arguments to your system and to get return value from your system.
def f(x):
    x0,x1 = x[:,0],x[:,1]
    fx = (x0 - 10) * (x0 - 10) + (x1 - 5) * (x1 - 5)
    #(10, 5) is best value.
    print('x=', x, 'fx=', fx)
    return fx

# parameter boundary
# Discrete value is available. Please use 'discrete' instead of 'continuous'.
bounds = [{'name': 'x0', 'type': 'continuous', 'domain': (5,15)},{'name': 'x1', 'type': 'continuous', 'domain': (0, 10)}]

# instance class
myBopt = GPyOpt.methods.BayesianOptimization(f=f, domain=bounds)
```

```python
# optimization
myBopt.run_optimization(max_iter=30)
myBopt.plot_acquisition()
myBopt.plot_convergence()
```

```python
# show optimized parameters
print(myBopt.x_opt)
print(myBopt.fx_opt)
```
